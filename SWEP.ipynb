# Student Performance Analysis

This project analyzes factors affecting student academic performance using the Portuguese student dataset.

Dataset Source: UCI Machine Learning Repository

Tools Used:
- Python
- Pandas
- Seaborn
- Matplotlib
- Jupyter Notebook

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv("student-por.csv")

df.head()
```

Does study time affect final grade?

```python
sns.boxplot(x='studytime', y='G3', data=df)

plt.title("Study Time vs Final Grade")
plt.show()
```

Do absences affect grades?

```python
sns.scatterplot(x='absences', y='G3', data=df)

plt.title("Absences vs Final Grade")
plt.show()
```

Does alcohol consumption affect grades?

```python
sns.boxplot(x='Walc', y='G3', data=df)

plt.title("Weekend Alcohol Consumption vs Final Grade")
plt.show()
```

Male vs Female Performance

```python
sns.boxplot(x='sex', y='G3', data=df)

plt.title("Gender vs Final Grade")
plt.show()
```

## Key Findings

1. Students with higher study time tend to achieve better grades.
2. Increased absences negatively affect student performance.
3. High alcohol consumption slightly correlates with lower grades.
4. Female students show slightly higher average performance than male students.
